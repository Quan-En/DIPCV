# 4. Visualization

```python
def GetPlotDF(arr_list):
    result_df = pd.DataFrame(np.column_stack([np.concatenate(arr_list[1]), np.row_stack(arr_list[0])]))
    result_df = result_df.sort_values(0, ascending=False).reset_index(drop=True)
    result_df[0] = result_df[0].astype(int)
    return result_df
```

# 5. Classifier

```python
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score, confusion_matrix
```

## 5.1. Balance data

```python
def GetBalanceTrainArr(arr_list):
    x_arr_list, y_arr_list = arr_list
    real_image_num = len(y_arr_list[2])
    each_type_num = real_image_num // 4
    for i in [0,1,3,4]:
        x_arr_list[i] = x_arr_list[i][:each_type_num,:]
        y_arr_list[i] = y_arr_list[i][:each_type_num]
    
    x_arr = np.row_stack(x_arr_list)
    y_arr = np.concatenate(y_arr_list)
    y_arr[y_arr != 0] = 1
    return x_arr, y_arr
```

```python
train_x_0, train_y_0 = GetBalanceTrainArr(train_0)
train_x_1, train_y_1 = GetBalanceTrainArr(train_1)
```

## 5.2. Logistic regression

```python
l_reg = LogisticRegression(penalty='l2', max_iter=1000)
l_reg.fit(train_x_0, train_y_0)
```

```python
train_acc_logistic = round(accuracy_score(train_y_0, l_reg.predict(train_x_0)), 4)
test_acc_logistic = round(accuracy_score(train_y_1, l_reg.predict(train_x_1)), 4)

print("Training ACC: ", train_acc_logistic), \
print("Testing ACC: ", test_acc_logistic)
```

```python
print("Training Confusion matrix: \n", confusion_matrix(train_y_0, l_reg.predict(train_x_0)))
print("Testing Confusion matrix: \n", confusion_matrix(train_y_1, l_reg.predict(train_x_1)))
```
